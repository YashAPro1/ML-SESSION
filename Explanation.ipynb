### Linear Regression

- Linear Regression is a statistical method to model the relationship between a dependent variable and one or more independent variables.
- Predict the value of the dependent variable based on the values of independent variables.
- Equation: 

#### Linear Regression for One Variable


- Equation that we use for LR is y=mx+b+e ---> for simple linear regression with one independent variable.
- Y is the dependent variable (the variable you are trying to predict).

#### Linear Regression for multiple variable



##### Y=0+1X1+2X2++nXn+
- Y is the dependent variable (the variable you are trying to predict).
- 0 is the y-intercept (the constant term).
- 1,2,,n are the coefficients for the independent variables X1,X2,,Xn, respectively.
-  represents the error term, which accounts for unobserved factors or random noise.

## Logistic Regression as Classification model

- Logistic Regression is a classification algorithm used for predicting the probability of an observation belonging to a particular class.
- Binary classification problems like spam detection, medical diagnosis.



- Logistic Regression uses the sigmoid function to squash predicted values between 0 and 1.
- Iteratively adjusts weights () to maximize the likelihood of the observed data.
- A threshold is set (often 0.5) to classify observations into classes (0 or 1).

# Introduction to Unsupervised Learning

- Learning without labeled output.
- Extracting patterns, relationships, and structures from data.



## Clustering in Unsupervised Learning

- Grouping similar data points together.
- Algorithms: K-Means, Hierarchical Clustering, DBSCAN.

### KMeans Algorithm

- K-Means is an unsupervised machine learning algorithm for clustering data into k clusters based on similarity.
- Minimize the sum of squared distances within clusters.
##### Algorithm Steps:
    - Initialize centroids.
    - Assign each data point to the nearest centroid.
    - Update centroids based on the mean of points assigned to them.
    - Repeat steps 2 and 3 until convergence.
